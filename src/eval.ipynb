```python
params = read_config(env_settings.CONFIG)
```

```python
processor = DataHandler(opt=params["dataset"], mode='inference_on_val')
chexpert_data_module = ChexpertDataModule(opt=params['dataset'], processor=processor)
chexpert_data_module.setup('test')
```

```python
test_dataloader = chexpert_data_module.test_dataloader()
```

```python
len(test_dataloader)
```

```python
model = CycleGAN.load_from_checkpoint(model_path)
```

```python
model.to('cuda:0')
model.eval()
```

```python
gt_reports = []
pred_reports = []
```

```python
from tqdm import tqdm
for i, batch in tqdm(enumerate(test_dataloader)):
    img, label = batch['target'], batch['report']
    pred_label = model(img)
    pred_label = torch.sigmoid(pred_label)
    
    gt_reports.append(label.cpu())
    pred_reports.append(pred_label.cpu())
```

```python
# Assuming gt_reports and pred_reports are lists of tensors, concatenate them into single tensors
gt_reports = [x.detach().cpu() for x in gt_reports]
pred_reports = [x.detach().cpu() for x in pred_reports]
```

```python
import torch
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import accuracy_score, hamming_loss
import numpy as np

# CheXpert labels for reference
chexpert_labels = [
    'Atelectasis',
    'Cardiomegaly', 
    'Consolidation', 
    'Edema', 
    'Enlarged Cardiomediastinum',
    'Fracture',
    'Lung Lesion',
    'Lung Opacity',
    'No Finding',
    'Pleural Effusion',
    'Pleural Other',
    'Pneumonia',
    'Pneumothorax'
]

# Concatenate tensors
gt_reports_tensor = torch.cat(gt_reports, dim=0)
pred_reports_tensor_raw = torch.cat(pred_reports, dim=0)
pred_reports_tensor_0_1 = torch.where(pred_reports_tensor_raw > 0.5, 1, 0)

# Calculate accuracy for each label
accuracy_per_label = (gt_reports_tensor == pred_reports_tensor_0_1).float().mean(dim=0)

# Calculate mean accuracy across all labels
mean_accuracy = accuracy_per_label.mean().item()  # Convert to Python float for printing

# Convert tensors to numpy arrays for sklearn metrics
gt_reports_np = gt_reports_tensor.numpy()
pred_reports_np = pred_reports_tensor_0_1.numpy()

# Calculate precision, recall, and F1 score for each label
precision = precision_score(gt_reports_np, pred_reports_np, average=None)
recall = recall_score(gt_reports_np, pred_reports_np, average=None)
f1 = f1_score(gt_reports_np, pred_reports_np, average=None)

# Display metrics per label
metrics_per_label = {label: {'Precision': p, 'Recall': r, 'F1 Score': f} 
                     for label, p, r, f in zip(chexpert_labels, precision, recall, f1)}
for label, metrics in metrics_per_label.items():
    print(f"{label}: {metrics}")

# Calculate macro and micro metrics
precision_macro = precision_score(gt_reports_np, pred_reports_np, average='macro')
recall_macro = recall_score(gt_reports_np, pred_reports_np, average='macro')
f1_macro = f1_score(gt_reports_np, pred_reports_np, average='macro')

precision_micro = precision_score(gt_reports_np, pred_reports_np, average='micro')
recall_micro = recall_score(gt_reports_np, pred_reports_np, average='micro')
f1_micro = f1_score(gt_reports_np, pred_reports_np, average='micro')

# Calculate subset accuracy and hamming loss
exact_match_ratio = accuracy_score(gt_reports_np, pred_reports_np)
hamming_loss_value = hamming_loss(gt_reports_np, pred_reports_np)

# Calculate AUC-ROC for each label (use raw probabilities)
pred_probs_np = pred_reports_tensor_raw.numpy()  # This should be your sigmoid outputs as numpy arrays
auc_roc_per_label = roc_auc_score(gt_reports_np, pred_probs_np, average='macro', multi_class='ovr')

print(f"\nMean Accuracy: {mean_accuracy}")
print(f"Precision Macro: {precision_macro}")
print(f"Recall Macro: {recall_macro}")
print(f"F1 Score Macro: {f1_macro}")
print(f"Precision Micro: {precision_micro}")
print(f"Recall Micro: {recall_micro}")
print(f"F1 Score Micro: {f1_micro}")
print(f"Exact Match Ratio (Subset Accuracy): {exact_match_ratio}")
print(f"Hamming Loss: {hamming_loss_value}")
print(f"AUC-ROC per label (Macro): {auc_roc_per_label}")
```

```python
import numpy as np
from sklearn.metrics import f1_score

# Assuming pred_probs_tensor contains your model's probabilistic predictions for the validation set
pred_probs_np = pred_reports_tensor_raw
gt_labels_np = gt_reports_np

# Define a range of potential thresholds
thresholds = np.linspace(0, 1, num=100)

# Initialize the best score and optimal threshold variables
best_score = 0
optimal_threshold = 0.5

# Iterate over thresholds to find the best one
for threshold in thresholds:
    # Apply threshold
    pred_labels_thresholded = np.where(pred_probs_np > threshold, 1, 0)
    
    # Calculate the F1 score (or another metric of your choice)
    score = f1_score(gt_labels_np, pred_labels_thresholded, average='micro')  # 'macro' for unweighted mean
    
    # Update the best score and optimal threshold if the current score is better
    if score > best_score:
        best_score = score
        optimal_threshold = threshold

print(f"Optimal Threshold: {optimal_threshold}")
print(f"Best F1 Score: {best_score}")
```
